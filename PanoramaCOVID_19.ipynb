# Panorama da COVID-19

COVID-19 é um acrônimo para *coronavirus disease*, doença causada pelo coronavírus Sars-CoV-2. Trata-se de um vírus respiratório, cuja transmissão se da por gotículas expelidas por meio da tosse de pessoas infectadas. A doença, apesar de se apresentar de forma branda na maioria dos casos, assolou o mundo com 4 milhões de mortos, e toda essa trágica situação despertou questionamentos interessantes sobre como a doença atingiu os países e fatores que podem agravar os quadros.

## Obtenção dos Dados
O dataset disponível é fornecido pela Our World in Data. Organização que disponibiliza, de forma digital, estudos sobre questões que assolam o mundo atual. O site, criado por Max Roser lida, desde 2011, com questões como a fome, pobreza, guerras entre outras questões humanitárias, com o objetivo de levar as análises para as pessoas de forma simples e didática, fazendo com que a população reflita sobre essas questões. Com o início da pandemia da Covid-19 em 2020, o *Our World in Data* se dedicou profundamente a fornecer dados confiáveis sobre a doença no mundo e hoje, tem um dos, se não o melhor dataset disponível para análises. 

### **Fontes**
Os dados disponíveis nos repositórios, são alocados pelo *Our World in Data*, mas fornecidos por diversas organizações, veja a sepração:

* Casos e Mortes confirmados: Repositório de dados do centro de Ciência e Engenharia de sistemas da universidade John Hopkins (EUA).

* Hospitalizações: Europa - Centro Europeu de Controle e prevensão de doenças
Reino Unido, Estados Unidos, Canadá, Israel e Algéria - Governo local. **Outros países não constam dados para essa variável.**

* Vacinação: Relatórios oficiais dos governos
* Outras variáveis: ONU, Banco Mundial, entre outras (especificadas ao uso)


### **Dicionário de Variáveis**
* iso_code - caracterização padronizada dos países, mais em: https://pt.wikipedia.org/wiki/ISO_3166-1_alfa-3
* continent: continente
* location: Coordenadas geográficas
* date: Data
#### **Dados gerais da doença**
* total_cases: Total de casos registrados de Covid-19 
* new_cases: Novos casos de Covid-19 no dia
* new_cases_smoothed: Média móvel de novos casos nos últimos 7 dias
* total_deaths: Total de mortes atribuídas à Covid-19
* new_deaths: Novas mortes atribuídas à Covid-19 no dia
* new_deaths_smoothed: Média móvel de novas mortes atribuídas à Covid-19 nos últimos 7 dias
* total_cases_per_million: Total de casos a cada 1.000.000 de pessoas
* new_cases_per_million: Novos casos confirmados a cada 1.000.000 de pessoas
* new_cases_smoothed_per_million: Média móvel dos últimos 7 dias de novos casos a cada 1.000.000 de pessoas
* total_deaths_per_million: Total de mortes a cada 1.000.000 de pessoas
* new_deaths_per_million: Novas mortes confirmadas a cada 1.000.000 de pessoas
* new_deaths_smoothed_per_million: Média móvel dos últimos 7 dias de novas mortes a cada 1.000.000 de pessoas
* reproduction_rate: Estimativa da taxa de repordução efetiva da Covid-19. Para mais informações, veja:  https://github.com/crondonm/TrackingR/tree/main/Estimates-Database
#### **Hospitalizações**
* icu_patients: Pacientes em UTI com a Covid-19 no dia
* icu_patients_per_million: Pacientes em UTI com a Covid-19 no dia por milhão de pessoas
* hosp_patientes: Pacientes internados com a Covid-19 no dia
* hosp_patients_per_million: Pacientes internados com a Covid-19 no dia por milhão de pessoas
* weekly_icu_admissions: Novos pacientes em UTI para a semana
* weekly_icu_admissions_per_million: Novos pacientes em UTI para a semana por milhão de pessoas
* weekly_hosp_admissions: Novos pacientes internados para a semana
* weekly_hosp_admissions_per_million: Novos pacientes internados para a semana por milhão de pessoas
#### **Testagem**
* total_tests: Total de testes realizados para Covid-19
* new_tests: Novos testes realizados de Covid-19 (só é calculado para dias úteis)
* total_tests_per_thousand: total de testes realizados a cada 1.000 pessoas
* new_tests_per_thousand: novos testes realizados a cada 1.000 pessoas
* new_tests_smoothed: média móvel de novos testes realizados. **Tratamento diferente para países que não reportam testes diariamente, consulte: https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-codebook.csv**
* new_tests_smoothed_per_thousand: média móvel de novos testes realizados nos últimos 7 dias a cada 1.000 pessoas
* positive_tests: parcela dos testes realizados que são positivos, dada média simples dos últimos 7 dias corridos
* tests_per_case: quantos testes foram necessários para detectar um caso, dada média simples dos últimos 7 dias corridos
* tests_units: como os dados de teste foram reportados
#### **Vacinação**
* total_vaccinations: Total de doses de vacina aplicadas
* people_vaccinated: Total de pessoas que receberam pelo menos uma dose da vacina
* people_fully_vaccinated: total das pessoas com cobertura vacinal completa
* new_vaccinations: novas doses administradas (só para dias úteis)
* new_vaccinations_smoothed: média móvel de novas doses administradas. ****Tratamento diferente para países que não reportam vacinação diariamente, consulte: https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-codebook.csv**
* total_vaccinations_per_hundred: total de doses administradas a cada 100 habitantes
* people_fully_vaccinated_per_hundred: total de pessoas com cobertura vacinal completa a cada 100 habitantes
* new_vaccinations_smoothed_per_million: Média móvel de novas doses administradas a cada 1 milhão de habitantes.
#### **Outras variáveis**
* stringency_index: mede o quanto o país tomou medidas de contingência contra a covid-19. Variando de 0 (sem restrições) a 100(mais restritos). Veja mais em: https://ourworldindata.org/grapher/covid-stringency-index?time=2021-08-12
* population: população em 2020
* population_density: densidade demográfica
* median_age: projeção da mediana da idade pela ONU em 2020
* aged_65_older: parcela da população com 65 anos ou mais
* aged_70_older: parcela da população que tinha 70 anos ou mais em 2015
* gdp_per_capita: PIB per capita
* extreme_poverty: parcela da população vivendo em condições de extrema pobreza
* cardiovasc_death_rate: proporção de mortes por doenças cardiovasculares 2017.
* diabetes_prevalence: % da população entre 20 e 79 anos com diabetes
* female_smokers: parcela de mulheres fumantes
* male_smokers: parcela de homens fumantes
* handwashing_facilities: parcela da população com acesso a estrutura para higienização das mãos
* hospital_beds_per_thousand: leitos de hospital a cada 1000 habitantes.
* life_expectancy: expectativa de vida
* human_development_index: IDH
* excess_mortality: mortalidades em excesso de acordo com P-score. Consulte: https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality

## **A pandemia**
### O início da pandemia
Desde janeiro de 2020, comentava-se sobre uma nova doença respiratória que surgia na província de Wuhan na China, porém, sem alarmismo por parte do resto do mundo, que imaginava que pudesse ser mais uma epidemia local. Nessa mesma época, casos isolados foram detectados na Europa, mas o primeiro sinal de alerta para o mundo veio com o *outbreak* da Covid no norte da Itália, e o governo do país decidiu, no dia 9 de março de 2020, decretar lockdown no país. Vamos analisar os casos e mortes por Covid em março na Itália.

```
df_italia = df.loc[df.location == "Italy"].copy()
gridsize = (2, 2)

fig = plt.figure(figsize=(15,10))
# Plot de dados Itália Março

ax1 = plt.subplot2grid(gridsize, (0,0))
ax2 = plt.subplot2grid(gridsize, (0,1))
ax3 = plt.subplot2grid(gridsize, (1,0))
ax4 = plt.subplot2grid(gridsize, (1,1))

ax1 = df_italia.loc['2020-03-01':'2020-03-31'].new_cases.plot(ax=ax1, c='#013220')
ax2 = df_italia.loc['2020-03-01':'2020-03-31'].new_deaths.plot(ax=ax2, c='#013220')
ax3 = df_italia.loc['2020-03-01':'2020-03-31'].total_cases.plot(ax=ax3, c='red')
ax4 = df_italia.loc['2020-03-01':'2020-03-31'].total_deaths.plot(ax=ax4, c='red')

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('')
ax4.set_xlabel('')

ax1.set_title('Novos casos diários de Covid-19 em Março na Itália')
ax2.set_title('Novas mortes diárias devido à Covid-19 em Março na Itália')
ax3.set_title('Total de casos de Covid-19 em Março na Itália')
ax4.set_title('Total de mortes por Covid-19 em Março na Itália')


plt.tight_layout()
```

```
casos_marco = df_italia.loc['2020-03-01':'2020-03-31'].new_cases.sum()
mortes_marco = df_italia.loc['2020-03-01':'2020-03-31'].new_deaths.sum()

print("A Itália apresentou {:.0f} casos e {:.0f} mortes devido à Covid-19 em Março de 2020".format(casos_marco, mortes_marco))

 
```

### Alta de casos e posicionamento da OMS

#### Alta de casos nos continentes

```
df_continents = df.loc[:, ['continent', 'location', 'total_cases']]
df_continents = df_continents.loc['2020-03-01':'2020-03-31']
df_continents = df_continents.groupby('continent')

# criando grid com os casos de covid em cada pais
gridsize_cont = (2,3)
fig_cont = plt.figure(figsize=(15, 10)) 

ax1 = plt.subplot2grid(gridsize_cont, (0,0))
ax2 = plt.subplot2grid(gridsize_cont, (0,1))
ax3 = plt.subplot2grid(gridsize_cont, (0,2))
ax4 = plt.subplot2grid(gridsize_cont, (1,0))
ax5 = plt.subplot2grid(gridsize_cont, (1,1))
ax6 = plt.subplot2grid(gridsize_cont, (1,2))

df_as = df_continents.get_group('Asia').copy()
df_as.drop(['continent', 'location'], axis=1, inplace=True)
ax1 = df_as.groupby('date').sum().plot(ax=ax1, c='k')
ax1.set_xlabel('')
ax1.set_title('Total de casos de Covid na Ásia em Março')
ax1.legend(['Total de casos'], loc='upper left')

df_eu = df_continents.get_group('Europe').copy()
df_eu.drop(['continent', 'location'], axis=1, inplace=True)
ax2 = df_eu.groupby('date').sum().plot(ax=ax2, c='b')
ax2.set_xlabel('')
ax2.set_title('Total de casos de Covid na Europa em Março')
ax2.legend(['Total de casos'], loc='upper left')

df_na = df_continents.get_group('North America').copy()
df_na.drop(['continent', 'location'], axis=1, inplace=True)
ax3 = df_na.groupby('date').sum().plot(ax=ax3, c='r')
ax3.set_xlabel('')
ax3.set_title('Total de casos de Covid na América do Norte em Março')
ax3.legend(['Total de casos'], loc='upper left')

df_sa = df_continents.get_group('South America').copy()
df_sa.drop(['continent', 'location'], axis=1, inplace=True)
ax4 = df_sa.groupby('date').sum().plot(ax=ax4, c='c')
ax4.set_xlabel('')
ax4.set_title('Total de casos de Covid na América do Sul em Março')
ax4.legend(['Total de casos'], loc='upper left')

df_af = df_continents.get_group('Africa').copy()
df_af.drop(['continent', 'location'], axis=1, inplace=True)
ax5 = df_af.groupby('date').sum().plot(ax=ax5, c='m')
ax5.set_xlabel('')
ax5.set_title('Total de casos de Covid na África em Março')
ax5.legend(['Total de casos'], loc='upper left')

df_oc = df_continents.get_group('Oceania').copy()
df_oc.drop(['continent', 'location'], axis=1, inplace=True)
ax6 = df_oc.groupby('date').sum().plot(ax=ax6, c='y')
ax6.set_xlabel('')
ax6.set_title('Total de casos de Covid na Oceania em Março')
ax6.legend(['Total de casos', 'Decreto'], loc='upper left')

ax1.vlines('2020-03-11', 80000, 180000, linestyles='dotted', colors='red')
ax2.vlines('2020-03-11', -1000, 450000, linestyles='dotted', colors='red')
ax3.vlines('2020-03-11', 0, 200000, linestyles='dotted', colors='black')
ax4.vlines('2020-03-11', 0, 14000, linestyles='dotted', colors='red')
ax5.vlines('2020-03-11', 0, 6000, linestyles='dotted', colors='red')
ax6.vlines('2020-03-11', 0, 5500, linestyles='dotted', colors='red')


fig_cont.tight_layout()


```

### A variação do impacto da COVID-19 nos países de acordo com fatores socioeconômicos

#### PIB per capita

Como foi dito anteriormente, já vivemos mais de um ano de pandemia e essa vastidão de dados nos permite analisar como a Covid pode atingir países em diferentes intensidades, baseado em indicadores socioeconômicos.

```
df_hdi= df.loc[:, ['gdp_per_capita', 'total_deaths_per_million', 'location']]
df_hdi_atual = df_hdi.loc['2021-08-14'].copy()


df_hdi_atual.reset_index(inplace=True)
df_hdi_atual.drop('date', axis=1, inplace=True)
df_hdi_atual.dropna(axis=0, inplace=True)

fig, ax = plt.subplots(figsize=(15,10));
fig= plt.figure();

ax = df_hdi_atual.plot(kind='scatter', x='total_deaths_per_million', y='gdp_per_capita', ax=ax);
ax.set_ylabel("PIB per capita")
ax.set_xlabel("Total de mortes por milhão");

fig.tight_layout();


```

Esse gráfico analisa o total de mortes por milhão e o *PIB per capita* dos países. O *PIB per capita* é um índice socioeconômico que nos mostra o desenvolvimento econômico de um país. Ao comparar a renda total do país (PIB) com a população, é possível entender com quanto dinheiro a população vive em média. **Nosso gráfico mostrou que há uma tendência de aumento no número de mortes por milhão conforme diminuição do *PIB per capita*.**
